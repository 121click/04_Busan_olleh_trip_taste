```python
df = pd.read_csv('/content/drive/MyDrive/Data/blog_content_final.csv', encoding='utf-8-sig')
df
```

```python
# 전체 content에서 가장 글자수가 긴 content를 찾아서 그 글자수를 출력
# max_len = 0
# for i in range(len(df)):
#     if max_len < len(df['content'][i]):
#         max_len = len(df['content'][i])
# print(max_len)

# 7014990
```

```python
# df의 content를 100만자 단위로 content_1 에서 content_8까지 나누어 컬럼에 저장

df['content_1'] = df['content'].str[:1000000]
df['content_2'] = df['content'].str[1000000:2000000]
df['content_3'] = df['content'].str[2000000:3000000]
df['content_4'] = df['content'].str[3000000:4000000]
df['content_5'] = df['content'].str[4000000:5000000]
df['content_6'] = df['content'].str[5000000:6000000]
df['content_7'] = df['content'].str[6000000:7000000]
df['content_8'] = df['content'].str[7000000:8000000]
df
```

```python
# Keyword 기준으로 100개 단위 행으로 나누어 저장

df_100 = df.iloc[:100]
df_200 = df.iloc[100:200]
df_300 = df.iloc[200:300]
df_400 = df.iloc[300:400]
df_500 = df.iloc[400:500]
df_600 = df.iloc[500:566]
```

```python
# df_100의 pos를 저장

df_100['mecab_pos_1'] = df_100['content_1'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_2'] = df_100['content_2'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_3'] = df_100['content_3'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_4'] = df_100['content_4'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_5'] = df_100['content_5'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_6'] = df_100['content_6'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_7'] = df_100['content_7'].apply(lambda x: mecab.pos(x))
df_100['mecab_pos_8'] = df_100['content_8'].apply(lambda x: mecab.pos(x))
df_100.tail()
```

```python
# mecab_pos_1에서 mecab_pos_8까지의 컬럼 mecab_pos_merge 컬럼에 저장

df_100['mecab_pos_merge'] = df_100['mecab_pos_1'] + df_100['mecab_pos_2'] + df_100['mecab_pos_3'] + df_100['mecab_pos_4'] + df_100['mecab_pos_5'] + df_100['mecab_pos_6'] + df_100['mecab_pos_7'] + df_100['mecab_pos_8']
df_100.tail()

```

```python
# mecab_pos_merge 컬럼에서 명사, 동사, 형용사를 추출하여 각각 mecab_noun, mecab_verb, mecab_adj 컬럼에 저장

df_100['mecab_noun'] = df_100['mecab_pos_merge'].apply(lambda x: [word for word, pos in x if pos in ['NNG', 'NNP']])
df_100['mecab_verb'] = df_100['mecab_pos_merge'].apply(lambda x: [word for word, pos in x if pos in ['VV']])
df_100['mecab_adj'] = df_100['mecab_pos_merge'].apply(lambda x: [word for word, pos in x if pos in ['VA']])
```

```python
# keyword, content, mecab_pos_merge, mecab_noun, mecab_verb, mecab_adj 컬럼만 추출하여 df_100_mecab에 저장하고 csv파일로 저장

df_100_mecab = df_100[['keyword', 'mecab_noun', 'mecab_verb', 'mecab_adj']]
df_100_mecab.to_csv('/content/drive/MyDrive/Data/df_100_mecab.csv', encoding='utf-8-sig')
```
