```python
df_keyword = df_keyword[['content']]
df_keyword.head()
```

```python
# 1행 남기고 전부 드랍

df_keyword = df_keyword.drop(df_keyword.index[1:])
df_keyword.head()
```

```python
# mecab으로 각 keyword 별 content 내의 형태소 분석

df_keyword['mecab_pos'] = df_keyword['content'].apply(lambda x: mecab.pos(x))
df_keyword.head()
```

```python
# mecab_pos에서 길이가 1보다 긴 것만 남기기

df_keyword['mecab_pos'] = df_keyword['mecab_pos'].apply(lambda x: [y for y in x if len(y[0]) > 1])
df_keyword.head()
```

```python
# mecab으로 명사인 형태소 뽑기

df_keyword['mecab_nn'] = df_keyword['content'].apply(lambda x: mecab.nouns(x))
df_keyword.head()
```

```python
# counter 불러오기

from collections import Counter
```

```python
# mecab_nn에서 빈도 세기 (value_counts)

df_keyword['mecab_nn_n'] = df_keyword['mecab_nn'].apply(lambda x: Counter(x))
df_keyword.head()
```

```python
# mecab_pos에서 형용사가 포함된 절을 뽑고 그 빈도도 같이 세기 (value_counts)

df_keyword['mecab_pos_va'] = df_keyword['mecab_pos'].apply(lambda x: [i[0] for i in x if i[1] == 'VA'])
df_keyword.head()
```

```python
# counter 이용해서 mecab_pos_va 내의 각 요소 세기

df_keyword['mecab_pos_va_count'] = df_keyword['mecab_pos_va'].apply(lambda x: Counter(x))
df_keyword.head()
```

```python
# mecab_pos에서 동사가 포함된 절을 뽑고 그 빈도도 같이 세기 (value_counts)

df_keyword['mecab_pos_v'] = df_keyword['mecab_pos'].apply(lambda x: [i[0] for i in x if i[1] == 'VV'])
df_keyword.head()
```

```python
df_keyword['mecab_pos_v_count'] = df_keyword['mecab_pos_v'].apply(lambda x: Counter(x))
df_keyword.head()
```

```python
# filter와 lambda 이용해서 mecab_pos_n에서 1글자보다 긴 명사만 뽑아서 mecab_pos_nn에 저장

df_keyword['mecab_pos_nn'] = df_keyword['mecab_pos_n'].apply(lambda x: filter(lambda y: len(y) > 1, x))

```
